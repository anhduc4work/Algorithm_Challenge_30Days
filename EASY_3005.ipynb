```python
class Solution(object):
    def maxFrequencyElements(self, nums):
        frequency = []
        max_frequency = 0
        new = []
        ans = 0
        while(nums):
            num = nums.pop()
            if num not in new:
                new.append(num)
                frequency.append(1)

                if max_frequency == 0:
                    max_frequency = 1
            else:
                for _, i in enumerate(new):
                    if num == i:
                        frequency[_] += 1
                        if frequency[_] > max_frequency:
                            max_frequency = frequency[_]
                    else:
                        pass

        for _ in frequency:
            if _ == max_frequency:
                ans += _
            else:
                pass
        return ans
solution = Solution()
```
