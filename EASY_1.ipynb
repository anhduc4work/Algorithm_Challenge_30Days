```python
class Solution(object):
    def twoSum(self, nums, target):
        LENGTH = len(nums)
        for i in range(LENGTH):
            first = nums[i]
            need = target - first
            for j in range(i+1, LENGTH):
                if nums[j] == need:
                    return [i, j]
                else:
                    pass
solution = Solution()
```

```python
solution.twoSum(nums = [2,7,11,15], target = 9)
```

```python
solution.twoSum(nums = [3,2,4], target = 6)
```
