```python
import time
start = time.time()
```

```python
class Solution(object):
    def evenOddBit(self, n):
        # N = n
        if n == 0:
            return [0, 0]
        else:
            # binary = ''
            even_count, odd_count = 0, 0
            count = 0
            while n > 0:
                remainder = n % 2
                # print(n, remainder)
                if remainder == 1:
                    if count % 2 == 1:
                        odd_count += 1
                    else:
                        even_count += 1
                else:
                    pass
                # binary = str(remainder) + binary
                n //= 2
                count +=1
            # print(binary)

            return [even_count, odd_count]
            
solution = Solution()
```

```python
solution.evenOddBit(50)
```
