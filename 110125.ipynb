```python
class Solution(object):

    def check(self, n, first = True):
        if n == 3:
            result = n * (n-1) // (n-2)
        elif n == 2:
            result = n * (n-1)
        elif n == 1:
            result = n
        else:
            pass

        if first:
            if n>3:
                return n * (n-1) // (n-2) + (n-3)
            else:
                return result
        else:
            if n>3:
                return -(n * (n-1) // (n-2)) + (n-3)
            else: 
                return -result

    def clumsy(self, n):
        if n <= 4:
            return self.check(n)
        else:
            result = self.check(n)
            for i in range(n-4, 0, -4):
                result += self.check(i, False) 
        return result

sol = Solution()
sol.clumsy(4)
```
