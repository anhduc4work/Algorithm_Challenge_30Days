```python
class Solution(object):
    def isGood(self, nums):
        """
        :type nums: List[int]
        :rtype: bool
        """
        ceil = max(nums)

        if len(nums) != ceil + 1:
            return False

        for _ in range(2):
            try:
                nums.remove(ceil)
            except ValueError:
                return False
        

        for num in range(1, ceil, 1):
            if num not in nums:
                return False        

        return True
```

```python
solution = Solution()
solution.isGood([2, 1, 3])
```

```python
solution.isGood([1, 3, 3, 2])
```

```python
solution.isGood([3, 4, 4, 1, 2, 1])
```
