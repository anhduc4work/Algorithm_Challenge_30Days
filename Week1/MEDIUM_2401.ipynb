```python
class Solution(object):
    def longestNiceSubarray(self, nums):
        LENGTH = len(nums)
        ans = 1
        i = 0
        while i < LENGTH:
            i_bit = self.bit_form(nums[i])
            record_value_one_index = [self.one_bit_index(i_bit)]
            max_length = 1
            flag = False
            for j in range (i+1, LENGTH):
                j_bit = self.bit_form(nums[j])
                j_record = self.one_bit_index(j_bit)

                if j_record:
                    for digit_1 in j_record:
                        for _, record in enumerate(record_value_one_index):
                            # print(_, record)
                            if digit_1 in record:
                                flag = True
                                i = j + (_+1) - len(record_value_one_index)
                                break
                            else:
                                pass
                        if flag: break
                    if flag == True:
                        break
                    else:
                        max_length += 1  
                        record_value_one_index += [j_record]
                else:
                    max_length += 1

            if max_length > ans:
                ans = max_length
            if flag:
                pass
            else:
                i += 1
        return ans
    def bit_form(self, num):
        if num == 1:
            return '01'
        elif num == 0:
            return '00'
        else:
            bit = ''
            while num:
                d = num % 2
                bit = str(d) + bit
                num //= 2
            return bit
    def one_bit_index(self, num):
        LENGTH = len(num)
        index = []
        for _, digit in enumerate(num):
            if digit == '1':
                index.append(_ - LENGTH)
        return index
solution = Solution()
```
