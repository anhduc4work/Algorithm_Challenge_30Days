```python
import time
start = time.time()
```

```python
class Solution(object):
    def countSymmetricIntegers(self, low, high):
        """
        Target: number with even number of digits + equal sum of first half and second half

        :type low: int
        :type high: int
        :rtype: int
        """
        ans = []
        for i in range(low, high + 1):
            # filter out odd n_digits
            n_digits = len(str(i))
            if (n_digits % 2 != 0):
                pass
            else:
                # print(i) #debug
                first_half = int(str(i)[:n_digits // 2])
                second_half = int(str(i)[n_digits // 2:])

                first_half_sum = 0
                second_half_sum = 0
                for _ in range(n_digits):
                    first_half_sum += first_half % 10
                    first_half = first_half // 10
                    
                    second_half_sum += second_half % 10
                    second_half = second_half // 10

                if (first_half_sum == second_half_sum):
                    ans.append(i)
        return len(ans), ans
```

```python
solution = Solution()
solution.countSymmetricIntegers(1, 100)
```

```python
solution.countSymmetricIntegers(1200, 1230)
```

```python
total = time.time() - start
print(f"Time taken: {total/60:.2f} mins")
```
