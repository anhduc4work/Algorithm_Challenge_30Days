```python
class Solution(object):
    def specialArray(self, nums):
        LENGTH = len(nums)

        sorted_nums = sorted(nums)
        if LENGTH == 0:
            return 0

        if sorted_nums[-LENGTH] >= LENGTH:
            return LENGTH

        for i in range(LENGTH-1, 0, -1):
            
            
            if sorted_nums[-i] >= i and sorted_nums[-i-1] < i:
                return i
            else: pass
        

        return -1

solution = Solution()
```

```python
solution.specialArray([3,5])
```

```python
solution.specialArray([0,0])
```
