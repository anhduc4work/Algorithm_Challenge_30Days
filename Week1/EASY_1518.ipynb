# Solution

```python
import time
start = time.time()
```

```python
class Solution(object):
    def numWaterBottles(self, numBottles, numExchange):
        """
        :type numBottles: int
        :type numExchange: int
        :rtype: int
        """
        total_drink = 0
        can_drink = numBottles
        # residual = 0
        while can_drink > 0:
            total_drink += can_drink              
            can_drink = numBottles // numExchange 
            numBottles = numBottles % numExchange + can_drink 
        return total_drink

solution =Solution()
```

```python
solution.numWaterBottles(9,3)
```

```python
solution.numWaterBottles(15,4)
```

# Performance

```python
total = time.time() - start
print(f"Time taken: {total/60:.2f} mins")
```
