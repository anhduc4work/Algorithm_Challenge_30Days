```python
import time
start =time.time()
```

```python
class Solution(object):
    def mostWordsFound(self, sentences):
        max_length = 0
        for sentence in sentences:
            word_counts = len(sentence.split(" "))
            if word_counts > max_length:
                max_length = word_counts
            else:
                pass
        return max_length
solution = Solution()
```

```python
solution.mostWordsFound(["alice and bob love leetcode", "i think so too", "this is great thanks very much"])
```

```python
solution.mostWordsFound(["please wait", "continue to fight", "continue to win"])
```

# Performance

```python
total = time.time() - start
print(f"Time taken: {total/60:.2f} mins")
```

# Best

```python
class Solution(object):
    def mostWordsFound(self, sentences):
        """
        :type sentences: List[str]
        :rtype: int
        """
        max = 0
        for i in sentences:
            if len(i.split()) > max:
                max = len(i.split())
        return max
```
