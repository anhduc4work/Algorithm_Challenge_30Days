# Solution

```python
import time
start = time.time()
```

```python
class Solution():
    def isBalanced(self, num):
        LENGTH = len(num)
        odd_sum, even_sum = 0, 0
        if LENGTH % 2 == 0:
            for _ in range(1, LENGTH+1, 2):
                odd_sum += int(num[_])
                even_sum += int(num[_-1])
        else:
            for _ in range(0, LENGTH-1, 2):
                odd_sum += int(num[_])
                even_sum += int(num[_+1])
            odd_sum += int(num[-1])
        if odd_sum == even_sum:
            return True
        else:
            return False
solution = Solution()
```

```python
solution.isBalanced('1234')
```

```python
solution.isBalanced('124322222222')
```

```python
total = time.time() - start
print(f"Time taken: {total/60:.2f} mins")

# about 20'
```
