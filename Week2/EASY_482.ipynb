```python
import time
start = time.time()
```

```python
class Solution(object):
    def licenseKeyFormatting(self, s, k):
        s = s[::-1]
        new = ''
        count = 0
        for digit in s:
            if digit != '-':
                count += 1
                new += digit

                if count % k == 0:
                    new += '-'
            else:
                pass
        return new[::-1].lstrip('-').upper()

solution = Solution()
```

```python
solution.licenseKeyFormatting( s = "5F3Z-2e-9-w", k = 4)
```

```python
solution.licenseKeyFormatting(s = "2-5g-3-J", k = 2)
```
