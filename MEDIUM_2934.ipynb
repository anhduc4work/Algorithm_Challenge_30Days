```python
import time
start = time.time()
```

```python
class Solution(object):
    def minOperations(self, nums1, nums2):
        last_nums1 = nums1.pop()
        last_nums2 = nums2.pop()

        length = len(nums1)
        if length == 0:
            return 0
        else:
            max_nums1 = max(nums1)
            max_nums2 = max(nums2)
            # print(last_nums1, last_nums2, max_nums1, max_nums2, length)
            

            if last_nums1 != last_nums2:
                smaller_last = min(last_nums1, last_nums2)
                bigger_last = max(last_nums1, last_nums2)

                # save larger number idx
                list_1, list_2 = [], []
                for _ in range(length):
                    if nums2[_] > smaller_last and nums1[_] > smaller_last:
                        return -1
                    elif nums1[_] > bigger_last or nums2[_] > bigger_last:
                        return -1
                    else:
                        if nums1[_] > smaller_last:
                            list_1.append(_)
                        if nums2[_] > smaller_last:
                            list_2.append(_)

                # Swap smaller last one to the list with less number larger than it
                count_1 = len(list_1)
                count_2 = len(list_2)
                # print(count_1, count_2)
                
                min_count= min(count_1, count_2)
                max_count = max(count_1, count_2)

                    
                if count_1 == count_2:
                    return count_1
                    
                else:
                    if count_1 < count_2 and smaller_last == last_nums1:
                        return min_count
                    elif count_1 < count_2 and smaller_last == last_nums2:
                        return min_count + 1
                    elif count_1 > count_2 and smaller_last == last_nums1:
                        return min_count + 1
                    elif count_1 > count_2 and smaller_last == last_nums2:
                        return min_count
                
            else:
                if max(max_nums1, max_nums2) > last_nums1:
                    return -1
                else:
                    return 0
                
```

```python
solution = Solution()
nums1 = [1,2,7]
nums2 = [4,5,3]
solution.minOperations(nums1, nums2)
```
